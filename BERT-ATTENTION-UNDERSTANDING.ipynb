```python
from transformers import BertModel, BertTokenizer
```

```python
from bertviz import head_view
```

```python
import torch
import pandas as pd
```

```python
#load vennila bert model
tokenizer  = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
model = BertModel.from_pretrained("bert-base-uncased")
```

```python
text = "My friend told me about the class and i loved it so far"
```

```python
len(model.encoder.layer)
```

```python
tokens = tokenizer.encode(text)
```

```python
inputs = torch.tensor(tokens).unsqueeze(0) #unsqueeze changes the shape from (20,) to (1,20)
```

```python
inputs
```

```python
attention = model(inputs, output_attentions = True)[2] # grab attention score from BERT
```

```python
final_attention = attention[-1].mean(1)[0] #average attention in last encoder
```

```python
attention_df = pd.DataFrame(final_attention.detach()).applymap(float).round(3)
```

```python
attention_df.columns = tokenizer.convert_ids_to_tokens(tokens)
attention_df.index = tokenizer.convert_ids_to_tokens(tokens)
```

```python
attention_df
```

```python
tokens_as_list = tokenizer.convert_ids_to_tokens(inputs[0])
head_view(attention,tokens_as_list)
```
